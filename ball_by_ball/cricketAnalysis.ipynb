```python
df[['wicket_type' , 'player_out']] = df[['wicket_type' , 'player_out']].fillna('N.O')
df[['target_runs' , 'target_overs' , 'runs_remaining' , 'balls_remaining' , 'required_run_rate' , 'pressure_index']] = df[['target_runs' , 'target_overs' , 'runs_remaining' , 'balls_remaining' , 'required_run_rate' , 'pressure_index']].fillna(0) 
```

```python
df.drop(columns= 'Unnamed: 0' , inplace=True)
```

```python
df.head()
```

```python
# Total Matches Played 
Total_matches_played = df.groupby('match_id')['match_id'].value_counts().shape[0]
```

```python
# Total Matches Won By Each Team 
dff = df.groupby('winner')['match_id'].value_counts().reset_index()
dff = dff.groupby('winner')['winner'].value_counts().sort_values(ascending=False).reset_index()
plt.barh(dff['winner'] , dff['count'])
plt.xlabel("Total Matches Won")
plt.ylabel("Teams")
plt.title("Total matches won by each team")
```

```python
# Most Wicket By Any Bowler 
a = df[df['wicket_type'] != 'N.O']
a.groupby('bowler')['bowler'].value_counts().sort_values(ascending=False)
```
